```python
# Checking for Outliers in the dataset
```

```python
flight.dtypes
```

```python
for col in flight.columns:
    if flight[col].dtypes!='object':
        sns.boxplot(flight[col])
        plt.xlabel(col)
```

```python
# Removing the outliers using IQR method
```

```python
Q1=flight['price'].quantile(0.25)
Q3=flight['price'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-(1.5*IQR)
upper_bound=Q3+(1.5*IQR)
```

```python
flight=flight[(flight['price']>=lower_bound) & (flight['price']<=upper_bound)]
```

```python
flight
```

```python
for col in flight.columns:
    if flight[col].dtypes!='object':
        sns.boxplot(flight[col])
        plt.xlabel(col)
```

```python
# Converting categorical column to Numerical for applying Machine Learning Algorithms
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for col in flight.columns:
    if flight[col].dtypes=='object':
        flight[col]=le.fit_transform(flight[col])
```

```python
flight
```

```python
# Splitting the datas into X and Y to apply ML models
X=flight.iloc[:,0:10]
y=flight.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
# Detecting Multicollinearity and removing the Multicollinear Features
from statsmodels.stats.outliers_influence import variance_inflation_factor
Vif=pd.DataFrame()
```

```python
Vif['feature']=X.columns
```

```python
Vif
```

```python
vif=[variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
```

```python
vif
```

```python
Vif['result']=vif
```

```python
Vif
```

```python
# Dropping the column which has Vif value > 5
X.drop(columns='flight',inplace=True)
```

```python
X
```

```python
vif=pd.DataFrame(X.columns,columns=['feature'])
```

```python
vif
```

```python
variance=[ variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
```

```python
vif['result']=variance
```

```python
vif
```

```python
# Splitting into Training and Testing Datas and Applying Linear Regression Algorithm to predict the future results
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
# Standardising to bring all the Values in the data frame to a single scale
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Applying Linear Regression Model
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
```

```python
# Training the dataset
lin.fit(X_train,y_train)
```

```python
# Testing using the testing set
y_pred=lin.predict(X_test)
```

```python
# Calculating the performace of the model
from sklearn.metrics import *
```

```python
# Evaluating the performance of the model using the R2_scores,MAE,RMSE
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
# The model can accurately predict 90% of the result correctly
```

```python
result=pd.DataFrame(y_test)
```

```python
result['predicted_price']=y_pred
```
