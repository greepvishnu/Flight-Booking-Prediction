```python
# Dayleft vs price
```

```python
sns.relplot(x=flight['days_left'],y=flight['price'],kind='line',data=flight,height=6,aspect=3)
```

```python
# From the line chart we can clearly say that as the no of days increases the flight price decreases
```

```python
#Average Price for each airlines
```

```python
avg_price=flight.groupby('airline')['price'].mean()
avg_price.plot.bar()
```

```python
# From the graph we can clearly saw that the avg price of Vistara is high and the avg price of Air Asia is the lowest
```

```python
# Stops vs Price
```

```python
sns.barplot(x=flight['stops'],y=flight['price'])
```

```python
# From the bar graph we can conclude that the price of one stops flight is higher when compared to zero stops or two or more
```

```python
# Business,Economy class,flight vs price
```
